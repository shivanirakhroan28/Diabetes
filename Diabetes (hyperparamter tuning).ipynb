```python
x= df.drop('Outcome',axis=1)
y= df['Outcome']
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
mn= MinMaxScaler()
```

```python
cols= x.columns
```

```python
x= mn.fit_transform(x)
```

```python
x= pd.DataFrame(x, columns=cols)
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test= train_test_split(x,y,random_state=42,shuffle=True,test_size=0.2,stratify=y)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import classification_report,accuracy_score,precision_score,recall_score,f1_score,roc_auc_score,roc_curve,auc
```

```python
models = {
    'KNeighbors Classifier': KNeighborsClassifier(n_neighbors=5,metric='manhattan'),
    'Logistic Regression': LogisticRegression(),
    'DecisionTreeClassifier': DecisionTreeClassifier(),
    'RandomForestClassifier': RandomForestClassifier(),
    'SVC': SVC()
}
```

```python
train_score={}
test_score={}
results=[]

for name, model in models.items():
    model.fit(x_train,y_train)
    y_pred= model.predict(x_test)
    
    train_score[name]= model.score(x_train,y_train)
    test_score[name]= model.score(x_test,y_test)
    accuracy= accuracy_score(y_pred,y_test)
    precision= precision_score(y_pred,y_test)
    recall= recall_score(y_pred,y_test)
    f1= f1_score(y_pred,y_test)
    
    results.append({
        'Model': name,
        'Train_score': train_score[name],
        'Test_score': test_score[name],
        'Accuracy': accuracy,
        'Precision': precision,
        'Recall': recall,
        'F1': f1
    })
result_df= pd.DataFrame(results)
```
